2. Analyze the average rating by country and region. Based on the insights and other relevant informamtion available in the data set, describe reputation of winery locations.

Step I need to take:
  
1. Data Preparation:

Load the dataset and filter countries with more than 50 reviews to ensure significant sample size.

2. Visualization:

Create a box plot to display the distribution of wine ratings by country, showing the median, interquartile range (IQR), and potential outliers.

3. Analysis:

Examine the spread of ratings, identify countries with consistent high-quality wines, and note any countries with wide variability or significant outliers.


```python
# Filter countries with more than 50 reviews for better visualization
countries_with_many_reviews = wine_data['country'].value_counts()[wine_data['country'].value_counts() > 50].index
filtered_data = wine_data[wine_data['country'].isin(countries_with_many_reviews)]

# Creating a box plot
plt.figure(figsize=(14, 8))
sns.boxplot(x='points', y='country', data=filtered_data, palette='Set2')
plt.title('Distribution of Wine Ratings by Country')
plt.xlabel('Wine Rating (Points)')
plt.ylabel('Country')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Explain the visual:

The box plot shows the distribution of wine ratings by country. Each box represents the middle 50% of the ratings (IQR), with the horizontal line inside the box showing the median rating. The whiskers extend to the lowest and highest ratings within 1.5 times the IQR, while points beyond this range are considered outliers.

Countries like Italy, France, and the US show high median ratings with relatively narrow IQRs, indicating consistent high-quality wines. Countries such as Brazil and Romania exhibit wider variability, suggesting a broader range of wine quality. Outliers in countries like Italy and Spain suggest some exceptionally high-rated wines.


3. Analyze wine varieties that consistently receive higher ratings across different regions. Based on the insights and other relevant information in the data set, describe popularity and consumer preference.

Step I will take:
1. Group the data by wine variety and region to calculate the average rating for each variety.
2. Identify wine varieties with consistently high ratings across multiple regions.

4. Analyze the popularity and consumer preferences based on the insights from the plot.

```python
# Calculate the average rating for each wine variety
average_rating_by_variety = wine_data.groupby('variety')['points'].mean().sort_values(ascending=False)

# Filter varieties with more than 100 reviews for significant data
varieties_with_many_reviews = wine_data['variety'].value_counts()[wine_data['variety'].value_counts() > 100].index
filtered_variety_data = wine_data[wine_data['variety'].isin(varieties_with_many_reviews)]

# Creating a box plot for top varieties
top_varieties = filtered_variety_data.groupby('variety')['points'].mean().sort_values(ascending=False).index[:10]
top_variety_data = filtered_variety_data[filtered_variety_data['variety'].isin(top_varieties)]

plt.figure(figsize=(14, 8))
sns.boxplot(x='points', y='variety', data=top_variety_data, palette='Set3')
plt.title('Distribution of Wine Ratings by Top Varieties')
plt.xlabel('Wine Rating (Points)')
plt.ylabel('Wine Variety')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Explanation of the visual:

1. Consistency Across Regions:
Narrow Boxes (e.g., Riesling, Champagne Blend): Suggest consistent high ratings, likely performing well across multiple regions.
Wider Boxes (e.g., Nebbiolo): Indicate more variability, possibly due to differing regional influences.
2. Popularity:
The varieties displayed are popular, frequently reviewed, and well-rated.
High outliers (e.g., Sagrantino, Port) may point to exceptional wines that drive consumer interest.

4. Analyze the wine prices by country, region, and variety. Based on the insights and other relevant information in the data set, describe the potential profitability and market positioning of the wines based on their countries, regions, and varieties.

Steps:
1. Data Analysis:
Analyze the average wine price by country, region, and variety.
Identify the top countries, regions, and varieties by average price.

2. Visualization:
Create bar plots to visualize the top countries, regions, and varieties by average price.

3. Insights:
Discuss the potential profitability and market positioning of wines based on pricing trends.


```python
#Drop rows with missing price values
wine_data = wine_data.dropna(subset=['price'])

#Analyze average wine price by country
average_price_by_country = wine_data.groupby('country')['price'].mean().sort_values(ascending=False)

#Analyze average wine price by region
average_price_by_region = wine_data.groupby('region_1')['price'].mean().sort_values(ascending=False).head(10)

#Analyze average wine price by variety
average_price_by_variety = wine_data.groupby('variety')['price'].mean().sort_values(ascending=False).head(10)

# Visualize the top 10 countries by average price
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_country.index[:10], y=average_price_by_country.values[:10])
plt.title('Top 10 Countries by Average Wine Price')
plt.xlabel('Country')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualize the top 10 regions by average price
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_region.index, y=average_price_by_region.values)
plt.title('Top 10 Regions by Average Wine Price')
plt.xlabel('Region')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualize the top 10 varieties by average price
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_variety.index, y=average_price_by_variety.values)
plt.title('Top 10 Varieties by Average Wine Price')
plt.xlabel('Wine Variety')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
