## Exercise 3:

```python
def checkSameElements(x,y):
 count = 0
 for i in x:
    if i in y:
      count = count + 1
 return count

t1 = (3, 'a', ['b', 2], 5)
t2 = (4, 5, ['b', 2])
checkSameElements(t1, t2)
```

```python
def build_map(list1, list2):
    result_map = {}
    seen_keys = set()
    list2_index = 0

    for key in list1:
        # Check if the key is mutable (list or dict)
        if isinstance(key, (list, dict)):
            continue  # Ignore mutable keys

        # If the key is a duplicate and already added, skip it
        if key in seen_keys:
            continue

        seen_keys.add(key)

        # Get corresponding value from list2, if available
        if list2_index < len(list2):
            result_map[key] = list2[list2_index]
            list2_index += 1
        else:
            result_map[key] = None  # Map extra keys to None if no value available

    # If list1 is longer than list2, map extra keys to None
    for key in list1:
        if key not in result_map and key not in seen_keys:
            result_map[key] = None

    return result_map

# Test Cases
# Test case 1
ls1 = [2, 5, 8, 5]
ls2 = ['we', 'me', 'he', 'they']
print(build_map(ls1, ls2))  # Expected: {2: 'we', 5: 'me', 8: 'he'}

# Test case 2
ls1 = ['a', 'b', [1, 2], 'c', 'b', {'e': 3}, 'd']
ls2 = [1, 2, 3, 4, 5, 6, 7]
print(build_map(ls1, ls2))  # Expected: {'a': 1, 'b': 2, 'c': 4, 'd': 7}

# Test case 3
ls1 = []
ls2 = ['we', 'me', 'he', 'they']
print(build_map(ls1, ls2))  # Expected: {}

# Test case 4
ls1 = [2, 5, 8, 5]
ls2 = ['we', 'me']
print(build_map(ls1, ls2))  # Expected: {2: 'we', 5: 'me', 8: None}

# Test case 5
ls1 = [2, 5, 5, 8]
ls2 = ['we', 'me', 'he', 'they', 'she']
print(build_map(ls1, ls2))  # Expected: {2: 'we', 5: 'me', 8: 'they'}

# Test case 6
ls1 = [2, 5, 5, 8]
ls2 = []
print(build_map(ls1, ls2))  # Expected: {2: None, 5: None, 8: None}
```
