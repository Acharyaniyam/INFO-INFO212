## Exercise 2:
Plot a histogram of the median_income column. What are the mean and median values of median_income? Write a brief discussion of your insights about the results in 100 words or less.

````python
#Plotting the histogram of the median_income column
plt.hist(housing['median_income'])
plt.xlabel('Median Income')
plt.title('Histogram of Median Income')
plt.show()

#Mean and median of median_income column
mean_median_income = housing['median_income'].mean()
median_median_income = housing['median_income'].median()

print("The mean of the median_income column is :", mean_median_income)
print("The median of the median_income column is :", median_median_income)


````

#The distribution in the histogram can be seen as skewed to the right. The mean is greater than the median of the median_income column hence, there are outliers in this dataset.


## Exercise 3:

Plot the distribution of median_house_value in box plot. What are the median, 25% percentile, 75% percentile, and mean of the median_house_value? Write a brief discussion of your insights about the results in 100 words or less.

````python
# Plotting the distribution of median_house_value in a box plot
plt.boxplot(housing['median_house_value'])
plt.ylabel('Median House Value')
plt.title('Box Plot of Median House Value')
plt.show()

# Calculating the median, 25% percentile, 75% percentile, and mean of median_house_value
median_house_value_median = housing['median_house_value'].median()
house_value_25 = housing['median_house_value'].quantile(0.25)
house_value_75 = housing['median_house_value'].quantile(0.75)
median_house_value_mean = housing['median_house_value'].mean()

print("The median of the median_house_value is:", median_house_value_median)
print("The 25% percentile of the median_house_value is:", house_value_25)
print("The 75% percentile of the median_house_value is:", house_value_75)
print("The mean of the median_house_value is:", median_house_value_mean)


````

# The box plot shows that the distribution of median_house_value is right-skewed with several outliers. The median is lower than the mean, indicating the presence of high-value houses that pull the mean upwards. The interquartile range (IQR) provides insight into the spread of the central data, while the outliers represent potential areas with exceptionally high housing prices.

## Exercise 4:

Plot the distribution of ocean_proximity into bar plot. Write a brief discussion of your insights about the results in 100 words or less.

````python
# Plot the distribution of ocean_proximity
housing['ocean_proximity'].value_counts().plot(kind='bar')
plt.xlabel('Ocean Proximity')
plt.title('Distribution of Ocean Proximity')
plt.show()
````

#The bar plot reveals that most houses are located near the ocean (<1H OCEAN) followed by INLAND. There are fewer houses near the bay and the island, indicating that majority of the houses are not located on the coast or islands.

## Exercise 5:

For each type of ocean_proximity, plot the median_house_value in box plot. Write a brief discussion of your insights about the results in 100 words or less.

````python

# Group data by ocean_proximity and plot box plots for median_house_value
housing.boxplot(column=['median_house_value'], by='ocean_proximity', figsize=(10, 6))
plt.title('Median House Value by Ocean Proximity')
plt.ylabel('Median House Value')
plt.xlabel('Ocean Proximity')
plt.show()
````

#Houses near the ocean tend to have higher median house values, particularly those less than 1 hour from the ocean. Island locations show a wide range of house values, potentially due to a smaller sample size. Inland locations have lower median house values compared to coastal areas.
