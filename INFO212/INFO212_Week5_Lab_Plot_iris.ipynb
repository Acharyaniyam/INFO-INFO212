[link text](https://)2. List the unique names of the species. Show the distribution of species.

```python
#get the unique values of species column
unique_species = data['species'].unique()
print("Unique species names:", unique_species)

#get the distribution of species
species_distribution = data['species'].value_counts()
print("\nDistribution of species:\n", species_distribution)
```

3.  Use seaborn.histplot() to plot histograms for each of the feature columns, colored by the species. You should have 4 histogram plots. What do you see from the plots? Explain the distributions against the species.

```python
# Iterate through the feature columns and create histograms for each
for feature in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
  plt.figure(figsize=(10, 6))
  sns.histplot(data=data, x=feature, hue='species', kde=True)  # kde=True adds kernel density estimate
  plt.title(f'Distribution of {feature} by Species')
  plt.xlabel(feature)
  plt.ylabel('Frequency')
  plt.show()

#Sepal Length: The distributions of sepal length for Iris-setosa are generally shorter compared to Iris-versicolor and Iris-virginica. Iris-versicolor and Iris-virginica have overlapping distributions, making it difficult to distinguish them based on sepal length alone.

#Sepal Width: Iris-setosa has a narrower sepal width distribution compared to the other two species. Iris-versicolor and Iris-virginica exhibit some overlap in their sepal width distributions.

#Petal Length: Petal length provides the most significant separation between species. Iris-setosa has much shorter petals compared to the other two species. Iris-versicolor and Iris-virginica have some overlap in their petal length distributions, but Iris-virginica tends to have slightly longer petals.

#Petal Width: Similar to petal length, petal width also offers a strong distinction between species. Iris-setosa has very narrow petals, while Iris-versicolor and Iris-virginica show a clear difference in petal width, with Iris-virginica having wider petals on average.

```

4. Scatter plot all pairs of different features and color the points with the species. You should have 6 scatter plots.

```python
# Create a list of feature pairs
feature_pairs = [
    ('sepal_length', 'sepal_width'),
    ('sepal_length', 'petal_length'),
    ('sepal_length', 'petal_width'),
    ('sepal_width', 'petal_length'),
    ('sepal_width', 'petal_width'),
    ('petal_length', 'petal_width')
]

# Loop through the feature pairs and create a scatter plot for each
for x_feature, y_feature in feature_pairs:
  plt.figure(figsize=(10, 4))  # Adjust figure size as needed
  sns.scatterplot(data=data, x=x_feature, y=y_feature, hue='species')
  plt.title(f'{x_feature} vs {y_feature}')
  plt.xlabel(x_feature)
  plt.ylabel(y_feature)
  plt.show()
```
