4. Define a function taking a DataFrame, Age column, and index as paramenter. The function categorizes Ages into Young, Middle_aged, and Senior category. Compute the Survived rate for each age group in descending order. Discuss insights you can draw from the results.

```python
def age_survival_rate(df, age_column="Age"):
    # Define age categories based on ranges
    categories = ['Young', 'Middle_aged', 'Senior']

    # Create 'AgeGroup' column with categorized age groups
    df['AgeGroup'] = pd.cut(df[age_column], bins=[0, 29, 59, float('inf')], labels=categories, right=False)

    # Compute survival rate for each age group and sort in descending order
    survival_rate = df.groupby('AgeGroup', observed=True)['Survived'].mean().sort_values(ascending=False)
    return survival_rate


survival_rates_by_age = age_survival_rate(df)
print(survival_rates_by_age)
```

Middle-aged passengers had the highest survival rate (41.93%), followed closely by young passengers (40.66%), while seniors had the lowest (25%), likely due to mobility challenges. This suggests younger and middle-aged groups had an advantage in survival odds.








5. Map the Embarked cities as 'S' to 'Southampton', 'C' to 'Cherbourg', 'Q' to 'Queenstown'. Compute the Survived rate by each embarked city using the mapping. Discuss insights you can draw from the results.

```python
# Map 'Embarked' codes to city names
df['Embarked'] = df['Embarked'].map({'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'})

# Compute the survival rate for each embarked city
survival_rate_by_city = df.groupby('Embarked')['Survived'].mean()
print(survival_rate_by_city)
```

Cherbourg passengers had the highest survival rate (55.36%) maybe because of the presence of more 1st Pclass passengers. Lower survival rates for Queenstown (38.96%) and Southampton (33.70%) suggest more 3rd Pclass passengers.








6. Define a function that computes the peak difference between the maximum and minimum values in an array. Group the data by Pclass and compute the peak difference of Ages for each Pclass group. Show the maximum, minimum, and the difference between max and min. Discuss insights you can draw from the results.

```python
# Define the function to compute peak difference
def peak_difference(arr):
    max_val = np.max(arr)
    min_val = np.min(arr)
    difference = max_val - min_val
    return max_val, min_val, difference

# Applied a lambda function on peak_difference function to where x is the table with each Pclass grouped to their age with droped the null values.
age_peak_difference_by_pclass = df.groupby('Pclass')['Age'].apply(lambda x: peak_difference(x.dropna()))

print(age_peak_difference_by_pclass)
```

First class has the widest age range (0.92 to 79.08), while third class also shows a broad range (0.42 to 73.58), indicating diverse age groups across classes. The wide age span in third class suggests many families or younger passengers, while first class likely includes older, wealthier people.
