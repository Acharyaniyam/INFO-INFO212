## Exercise 1:

Create 3 lists, `ls1`, `ls2`, and `ls3`, with the following values:
- `ls1` contains 10 random decimal numbers in the range of (0.00, 150.00), each of which with 2-digit precision.
- `ls2` contains 10 random integers in the range of (1, 500).
- `ls3` contains 10 random decimal numbers in the range of (0.05, 0.90), each of which with 2-digit precision and the last digit rounded to the nearest 0 or 5.

Show the content of the lists.

```python
ls1 = np.random.uniform(0.00, 150.00, 10)
ls2 = np.random.randint(1, 500, 10)

# Generate 10 random decimal numbers between 0.05 and 0.90, multiply by 20 to shift the decimal,
# round to the nearest integer (ensuring last digit is 0 or 5), then divide by 20 to scale back down,
# and round the result to 2 decimal places for precision.
ls3 = ls3 = np.round(np.round(np.random.uniform(0.05, 0.90, 10) * 20) / 20, 2)

print(ls1)
print(ls2)
print(ls3)
```

## Exercise 2:
- Use the 3 lists as 3 columns to create a numpy array `arr` which uses `ls1` as the first column, `ls2` as the second column, and `ls3` as the third column.
- Show the content and shape of the array.

```python
# Stack the three lists as columns to form a 2D numpy array
arr = np.column_stack((ls1, ls2, ls3))
print(arr)
```

## Exercise 3:
- For the third column of `arr`, compute a boolean array indicating whether the value is between [0.20, 0.50].
- Use the boolean array to extract the rows of `arr` such as the values of the third column are between [0.20, 0.50].

```python
third_column = arr[:, 2]  # Extract the third column of the array
boolean_array = (third_column >= 0.20) & (third_column <= 0.50)  # Create a boolean array based on the condition
extracted_rows = arr[boolean_array]  # Use the boolean array to extract the rows that meet the condition
print(extracted_rows)
```
