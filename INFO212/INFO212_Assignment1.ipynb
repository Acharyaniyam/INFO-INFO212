
## Qestion 3 [20 marks]:
Continue with the 'inflammation-data' directory. Given the following list of file names:
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv'],
```
write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](https://i.imgur.com/XXT2Jqw.png)

````python
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
````

````python
# Your code here



for filename in filenames:
  #Now loading the dataset
  inflammation = np.loadtxt("/content/drive/MyDrive/info212/inflammation-data/" + filename, delimiter = ',')

  #Getting things in an array
  daily_average = inflammation.mean(axis = 0)
  daily_max = inflammation.max(axis = 0)
  daily_min = inflammation.min(axis = 0)

  #Setting the scene for our visualization
  plt.figure(figsize = (10,3))

  #This following code gets the filename to be in the visual as well
  plt.suptitle(filename)

  #Plot for Average inflammation
  plt.subplot(1,3,1)
  plt.plot(daily_average)
  plt.ylabel("Average")

  #Plot for Maximum inflammation
  plt.subplot(1,3,2)
  plt.plot(daily_max)
  plt.ylabel("Maximum")

  #Plot for Minimum inflammation
  plt.subplot(1,3,3)
  plt.plot(daily_min)
  plt.ylabel("Minimum")

 # Display the plots
  plt.tight_layout()
  plt.show()


````
