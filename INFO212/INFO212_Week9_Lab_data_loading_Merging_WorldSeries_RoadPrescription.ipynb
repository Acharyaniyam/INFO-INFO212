3: List all the world series records that a team from Philadelphia played.  

---



```python
#First we need to get all the occurences that have the word 'Philadelphia' in the Wining and Losing team
philadelphia_teams_win = tables[1][tables[1]['Winning team'].str.contains('Philadelphia')]
philadelphia_teams_lose = tables[1][tables[1]['Losing team'].str.contains('Philadelphia')]
#Then concatinate them both to get everything.
philadelphia_teams = pd.concat([philadelphia_teams_win, philadelphia_teams_lose])
print("World Series records for Philadelphia teams:")
philadelphia_teams
```

4: What is the winning probability of the Phillies?

```python
#First to get total Phillies games
phillies_teams_win = tables[1][tables[1]['Winning team'].str.contains('Phillies')]
phillies_teams_lose = tables[1][tables[1]['Losing team'].str.contains('Phillies')]
phillies_total_games = len(phillies_teams_win) + len(phillies_teams_lose)
phillies_total_wins = len(phillies_teams_win)
# probability = total wins / total games played by the team 'Phillies'
phillies_win_prob = phillies_total_wins / phillies_total_games
print("Winning probability of the Phillies:", phillies_win_prob)
```

```python
# Using pandas' read_json funciton to complete the task.
file_path = '/content/drive/MyDrive/info212/roam_prescription_based_prediction.jsonl'
df = pd.read_json(file_path, lines=True)
print("\nShape of the DataFrame:", df.shape)
print("\nColumns of the DataFrame:\n", df.columns)
print("\nInfo of the DataFrame:\n")
print(df.info())
```

2: Extract the first 10,000 provider information as a DataFrame called `providers`. Set the index to be the npi.

```python
#Using the set_index function to keep the index as 'npi' column
providers = df.head(10000).set_index('npi')
providers
```

3: Extract first 10,000 prescription counts as a DataFrame called `pres`. Set the index to be the npi.

```python
#Same code as above but just specifying the column
pres = df.head(10000).set_index('npi')['cms_prescription_counts']
pres
```

4: Join `providers` with `pres` on npi. Extract the providers' info

*   List item
*   List item

who prescribed the most number of DIAZEPAM.

```python
# Parse the JSON-like prescription counts in `cms_prescription_counts`
pres_expanded = pres.apply(pd.Series)

# Join providers with the expanded prescription counts
combined = providers.join(pres_expanded)
combined

# Find the provider who prescribed the most DIAZEPAM
most_diazepam_npi = combined['DIAZEPAM'].idxmax()  # Get the NPI of the provider
provider_with_most_diazepam = combined.loc[most_diazepam_npi]  # Extract provider info

# Display the result
print("Provider Information for the most DIAZEPAM prescriptions:")
print(provider_with_most_diazepam)
```
