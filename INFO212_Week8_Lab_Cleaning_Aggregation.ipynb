1. Download and upload the nba-2016.csv. Create a DataFrame df from the data set. Show information of the DataFrame.

2. Does the dataset have mssing values? Compute the percentage of missing values overall.

```python
# Check if the dataset has missing values
has_missing_values = df.isnull().values.any()

# Calculate the percentage of missing values
missing_percentage = (df.isnull().sum().sum() / df.size) * 100

# Output the results
print(f"Does the dataset have missing values? {'Yes' if has_missing_values else 'No'}")
print(f"Percentage of missing values overall: {missing_percentage:.2f}%")
```

3. List the number and percentage of missing values for each row.

```python
# Calculate the number of missing values for each row
row_missing_count = df.isnull().sum(axis=1)

# Calculate the percentage of missing values for each row
row_missing_percentage = (row_missing_count / df.shape[1]) * 100

# Combine the results into a DataFrame
missing_values_per_row = pd.DataFrame({
    'Number of Missing Values': row_missing_count,
    'Percentage of Missing Values': row_missing_percentage
})

# Display the results
print(missing_values_per_row)
```

4. List the number and percentage of missing values for each column.

```python
# Calculate the number of missing values for each column
column_missing_count = df.isnull().sum()

# Calculate the percentage of missing values for each column
column_missing_percentage = (column_missing_count / df.shape[0]) * 100

# Combine the results into a DataFrame
missing_values_per_column = pd.DataFrame({
    'Number of Missing Values': column_missing_count,
    'Percentage of Missing Values': column_missing_percentage
})

# Display the results
print(missing_values_per_column)
```

6. Group by the Visitor/Neutral teams. Compute the aveage points (PTS) for each team playing as visiting team.

```python
# Group by the "Visitor/Neutral" column and compute the average points (PTS) for each visiting team
average_points_by_team = df.groupby('Visitor/Neutral')['PTS'].mean()

# Display the result
print(average_points_by_team)
```

7. Add a new column 'VisitorWon' which has True for the Visiting team won and False otherwise.

```python
# Add a new column 'VisitorWon' where True indicates the visiting team won
df['VisitorWon'] = df['PTS'] > df['PTS.1']

df.head()
```

8. Compute the rate of VisitorWon for each team playing away during the 2016 season, in descending order.

```python
#the rate of VisitorWon for each team for the away during 2016 season
visitor_wins_by_team = df.groupby('Visitor/Neutral')['VisitorWon'].mean().sort_values(ascending=False)


visitor_wins_by_team
```

9. Define a function 'top' that returns the top two rows based on a given col value. List each team's top 2 winning visiting games in which the team got top 2 highest points. Order the results by team names.

```python
# Define the 'top' function
def top(group, col, n=2):
    """Return the top n rows in the group based on the given column."""
    return group.nlargest(n, col)

# Filter for games where the visiting team won
winning_games = df[df['VisitorWon']]

# Group by visiting teams and apply the 'top' function
top_winning_games = (
    winning_games.groupby('Visitor/Neutral', group_keys=False)
    .apply(top, col='PTS', n=2)
)

# Sort the results by team names
top_winning_games = top_winning_games.sort_values(by='Visitor/Neutral')

# Display the results
top_winning_games
```

2. Forward fill missing values for each column. Ensure there is no missing value after the filling.

```python
# Forward fill missing values for each column
df_filled = df.ffill(axis=0)

# Check if there are any missing values remaining
missing_values_remaining = df_filled.isnull().values.any()

# Print the result
print(f"Are there any missing values left? {'Yes' if missing_values_remaining else 'No'}")
```

3. Create a pivot table containing the means of latest_price and ratings indexed by brand, model, and os. Put os on columns and also compute the margins.

```python
# Create a pivot table with means of 'latest_price' and 'ratings'
pivot_table = pd.pivot_table(
    df,
    values=['latest_price', 'ratings'],
    index=['brand', 'model'],
    columns='os',
    aggfunc='mean',
    margins=True
)

# Display the pivot table
pivot_table.head()
```
